```python
import os
import requests
from dotenv import load_dotenv

# Load the environment variables
load_dotenv()

# Get the API key from the .env file
apikey = os.getenv('ALPHA_VANTAGE_APIKEY')

if apikey is None:
    raise ValueError('APIKEY not found in .env file')

# Define the URL for the API request
url = 'https://www.alphavantage.co/query'

# Define the parameters for the API request
params = {
    'function': 'NEWS_SENTIMENT',
    'tickers': 'IBM',
    'sort': 'EARLIEST',
    'limit': '1000',
    'apikey': apikey
}

# Make the API request
r = requests.get(url, params=params)

# Check the status of the response
if r.status_code == 200:
    data = r.json()
    print(data)
else:
    print(f'Error: {r.status_code} - {r.text}')
```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Configure Selenium WebDriver
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")

# Path to your ChromeDriver
webdriver_service = Service()

# Initialize WebDriver
driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)

# URL to Forex.com news page
url = "https://www.forex.com/en-us/market-analysis/latest-research/"

# Load the page
driver.get(url)

# Allow the page to load completely
time.sleep(5)

# Function to load all news by scrolling


def load_all_news():
    while True:
        try:
            load_more_button = driver.find_element(
                By.CSS_SELECTOR, '.load-more-btn')
            load_more_button.click()
            time.sleep(3)
        except Exception as e:
            print("No more 'Load More' button found:", e)
            break


# Load all news
load_all_news()

# Extract news data
news_data = []
articles = driver.find_elements(By.CSS_SELECTOR, '.article')

for article in articles:
    title = article.find_element(By.CSS_SELECTOR, '.title').text
    date = article.find_element(By.CSS_SELECTOR, '.date').text
    summary = article.find_element(By.CSS_SELECTOR, '.summary').text
    link = article.find_element(By.CSS_SELECTOR, 'a').get_attribute('href')

    news_data.append({
        "title": title,
        "date": date,
        "summary": summary,
        "link": link
    })

# Close the driver
driver.quit()

# Convert to DataFrame
df = pd.DataFrame(news_data)

# Display the DataFrame
df.head()

```
