```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import fx_data_cleaner
from ta import add_all_ta_features

import warnings
warnings.filterwarnings('ignore')
```

```python
# preliminary data cleaning
usd_data = fx_data_cleaner.csv_to_columns('usdcad h1 2008 to 2024.csv')
eur_data = fx_data_cleaner.csv_to_columns('eurcad h1 2008 to 2024.csv')

# load analytical base tables

usdcad_h1 = pd.DataFrame(usd_data)
eurcad_h1 = pd.DataFrame(eur_data)

# amend datatypes for each column. combine date and time to datetime
usdcad_h1 = fx_data_cleaner.fx_data_type(usdcad_h1)
eurcad_h1 = fx_data_cleaner.fx_data_type(eurcad_h1)
```

```python
usdcad_h1.head()
```

```python
usdcad_h1_tech = add_all_ta_features(usdcad_h1,
                                    open='open',
                                    high='high',
                                    low='low',
                                    close='close',
                                    volume='tick_vol')

usdcad_h1_tech.head()
```

```python
msk = usdcad_h1_tech.drop(columns=['close', 'date_time']).columns
usdcad_h1_tech[msk] = usdcad_h1_tech[msk].shift()
usdcad_h1_tech
```

```python
usdcad_h1_tech.info()
```

```python
usdcad_h1_tech.columns
```

```python
corr_matrix = usdcad_h1_tech[['close', 'volume_adi', 'volume_obv', 'volume_cmf', 'volume_fi',
                             'volume_em', 'volume_sma_em', 'volume_vpt', 'volume_vwap', 'volume_mfi',
                             'volume_nvi']].corr()

volume_corr_matrix = corr_matrix[abs(corr_matrix) >= 0.5]['close'].dropna()
volume_corr_matrix
```

```python
corr_matrix = usdcad_h1_tech[['close', 'volatility_bbm', 'volatility_bbh', 'volatility_bbl',
                              'volatility_bbw', 'volatility_bbp', 'volatility_bbhi',
                              'volatility_bbli', 'volatility_kcc', 'volatility_kch', 'volatility_kcl',
                              'volatility_kcw', 'volatility_kcp', 'volatility_kchi',
                              'volatility_kcli', 'volatility_dcl', 'volatility_dch', 'volatility_dcm',
                              'volatility_dcw', 'volatility_dcp', 'volatility_atr', 'volatility_ui']].corr()

volatility_corr_matrix = corr_matrix[abs(corr_matrix) >= 0.5]['close'].dropna()
volatility_corr_matrix
```

```python
corr_matrix = usdcad_h1_tech[['close', 'trend_macd', 'trend_macd_signal', 'trend_macd_diff', 'trend_sma_fast',
                              'trend_sma_slow', 'trend_ema_fast', 'trend_ema_slow',
                              'trend_vortex_ind_pos', 'trend_vortex_ind_neg', 'trend_vortex_ind_diff',
                              'trend_trix', 'trend_mass_index', 'trend_dpo', 'trend_kst',
                              'trend_kst_sig', 'trend_kst_diff', 'trend_ichimoku_conv',
                              'trend_ichimoku_base', 'trend_ichimoku_a', 'trend_ichimoku_b',
                              'trend_stc', 'trend_adx', 'trend_adx_pos', 'trend_adx_neg', 'trend_cci',
                              'trend_visual_ichimoku_a', 'trend_visual_ichimoku_b', 'trend_aroon_up',
                              'trend_aroon_down', 'trend_aroon_ind', 'trend_psar_up',
                              'trend_psar_down', 'trend_psar_up_indicator',
                              'trend_psar_down_indicator']].corr()
trend_corr_matrix = corr_matrix[abs(corr_matrix)>=0.5]['close'].dropna()
trend_corr_matrix
```

```python
corr_matrix = usdcad_h1_tech[['close', 'momentum_rsi', 'momentum_stoch_rsi',
                              'momentum_stoch_rsi_k', 'momentum_stoch_rsi_d', 'momentum_tsi',
                              'momentum_uo', 'momentum_stoch', 'momentum_stoch_signal', 'momentum_wr',
                              'momentum_ao', 'momentum_roc', 'momentum_ppo', 'momentum_ppo_signal',
                              'momentum_ppo_hist', 'momentum_pvo', 'momentum_pvo_signal',
                              'momentum_pvo_hist', 'momentum_kama', 'others_dr', 'others_dlr',
                              'others_cr']].corr()
momentum_corr_matrix = corr_matrix[abs(corr_matrix) >= 0.5]['close'].dropna()
momentum_corr_matrix
```

```python
# Combine all related features from the different correlation matrices into one list
related_features = [
    (volume_corr_matrix.index),    
    (volatility_corr_matrix.index),
    (trend_corr_matrix.index),
    (momentum_corr_matrix.index)
]

# Flatten the list of lists into a single list of features
rel_feat = []
for feature_list in related_features:
    rel_feat.extend(feature_list)

# Remove duplicates by converting the list to a set and back to a list
rel_feat = (set(rel_feat))

# Display the list of related features
rel_feat
```

```python
x = sorted(rel_feat)
x.append('date_time')
x.remove('close')
usdcad_99k = usdcad_h1_tech[x]
# EXtract only columns with more than 99k valid instances
usdcad_99k = usdcad_99k.loc[:, usdcad_99k.notna().sum() > 99000]
usdcad_99k.info()
```

```python
usdcad_99k = usdcad_99k.fillna(0)
usdcad_99k
```

```python
usdcad_99k.to_csv('usdcad_h1_technical_data.csv')
```
